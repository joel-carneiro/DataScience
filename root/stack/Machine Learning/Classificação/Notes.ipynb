## Iris dataset

```python
from sklearn.model_selection import cross_val_predict
from sklearn import metrics
from sklearn import svm
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np
```

```python
%matplotlib notebook
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Iris.csv').iloc[:, 1:]
```

```python
df.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.scatterplot(x = 'SepalLengthCm', y = 'SepalWidthCm', data = df, hue='Species')
```

```python
sns.scatterplot(x = 'PetalLengthCm', y = 'PetalWidthCm', data = df, hue='Species')
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(['Species'], axis=1), df['Species'])
```

```python
model = svm.SVC()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
result = cross_val_predict(model, x_test, y_test, cv = 10)

print(metrics.classification_report(y_test, result))
```
