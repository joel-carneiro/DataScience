## Introdução

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Pares ordenados simples
x = np.arange(-10, 11)
y = x 

plt.plot(x, y);
plt.show()
```

```python
# Definindo labels
x = np.arange(-100, 100)
y = x ** 2

plt.plot(x, y);
plt.title('FUNÇÃO QUADRÁTICA')
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```

```python
# Definindo estilo da reta
x = np.arange(-5, 6)
y = x ** 2

plt.plot(x, y, 'o');
plt.show()
```

```python
# Dando zoom na imagem
x = np.arange(-100, 100)
y = x ** 3

plt.plot(x, y, 'o');
plt.axis([0, 50, 0, 2500])
plt.show()
```

```python
# Definindo parâmetros
x1 = np.arange(0, 100)

x2 = x1
y2 = x1 ** 1.2

x3 = x1
y3 = x1 ** 1.3

# Plotando todos os gráficos
plt.plot(x1, x1, 'r--', x2, y2, 'b--', x3, y3, 'g--')

plt.show()
```

## Plotando dados categóricos

```python
grupos = list('ABCDE')
valores = [4000, 2000, 1000, 500, 250]

# Gráfico de barras
plt.bar(grupos, valores)
plt.show()
```

```python
produto_A = np.random.rand(12)
produto_B = np.random.rand(12)

x1 = np.arange(len(produto_A))
x2 = [x + 0.25 for x in x1]

plt.style.use('seaborn')

plt.figure(figsize=(10, 5))

plt.bar(x1, produto_A, width=0.25, label='Produto A', color = '#034159')
plt.bar(x2, produto_B, width=0.25, label='Produto B', color = '#038C3E')

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

plt.xticks([x + 0.125 for x in range(len(produto_A))], meses);

plt.set_cmap('Spectral_r')

plt.show()
```

```python
grupos = list('ABCDE')
valores = [4000, 2000, 1000, 500, 250]

# Gráfico de barras
plt.barh(grupos, valores)
plt.show()
```

```python
vendas = [1, 2, 3, 4]
labels = ['SQL', 'JAVASCRIPT', 'JAVA', 'PYTHON']

colors = ['#034159', '#025951', '#02735E', '#038C3E']

explode = (0, 0, 0, 0)

plt.pie(vendas, labels = labels, explode=explode, colors=colors);

centre_circle = plt.Circle((0, 0), 0.60, fc='white')
fig = plt.gcf()
  
fig.gca().add_artist(centre_circle)
```
