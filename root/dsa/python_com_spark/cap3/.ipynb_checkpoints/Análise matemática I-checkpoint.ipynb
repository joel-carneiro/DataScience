```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import sympy as smp
%matplotlib inline
```

```python
def is_prime(number):
    predecessors = np.arange(2, number, 1)
    return not np.any((number % predecessors) == 0)
```

```python
df = pd.DataFrame({'index': np.arange(1000), 'number' : np.arange(1000)})
```

```python
df['square'] = df['number'] ** 2
df['cubic'] = df['number'] ** 3
df['diff_square'] = df['square'].diff()
df['diff_cubic'] = df['cubic'].diff()
df['is_prime'] = df['number'].apply(is_prime)
```

```python
df
```

```python
x = smp.Symbol('x')

f = x ** 2

f
```

```python
smp.diff(f)
```

```python
sns.lineplot(x='number', y='diff_square', data=df)
```

```python
g = x ** 3
g
```

```python
smp.diff(g)
```

```python
sns.lineplot(x='number', y='diff_cubic', data=df)
```

```python
primes = df.query('is_prime == True')
```

```python
primes['index'] = np.arange(len(primes))
```

```python
primes
```

```python
primes['diff_prime'] = primes['number'].diff()
```

```python
ax = sns.lineplot(x='index', y='diff_prime', data=primes)

ax.figure.set_size_inches(10, 5)
```
