```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(100)
y = x + np.random.choice(np.arange(-10, 10), 100, 100)
```

```python
plt.scatter(x, y)
```

```python
df = pd.DataFrame({'input' : x, 'output' : y})
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(df[['input']], df['output'])

model.score(df[['input']], df['output'])
```

```python
plt.scatter(x, y)
plt.plot(df['input'], model.predict(df[['input']]), 'r')
```
