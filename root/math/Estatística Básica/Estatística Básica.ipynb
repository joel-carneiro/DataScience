```python
altura_media = dataset["Altura"].mean()
```

```python
dataset["Altura"] - altura_media
```

```python
variancia = (dataset["Altura"] - altura_media) ** 2; variancia
```

```python
desvio_padrao = np.sqrt(variancia); desvio_padrao
```

```python
np.mean(desvio_padrao)
```
