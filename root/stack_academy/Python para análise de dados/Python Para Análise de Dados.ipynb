## 1.4 - Consulta de dados

```python
# Contar os valores unicos de uma Series
dataset['bedrooms'].unique()
```

```python
# Contar a aparição de cada dado em uma Series
pd.value_counts(dataset['bedrooms'])
```

```python
# Consultar imóveis com 3 Quartos
dataset.loc[dataset['bedrooms'] == 3]
```

```python
dataset.loc[(dataset['bedrooms'] == 3) & (dataset['bathrooms'] > 2)]
```

```python
dataset.sort_values(by='price', ascending=False)
```

```python
# Adicionando nova coluna
dataset['size'] = dataset['bedrooms'] * 20
```

```python
dataset
```

```python
values = dataset['price'].quantile([0.25, 0.5, 0.75]).tolist()
```

```python
def categorize(price):
    if price <= values[0]:
        return 'A'
    elif price <= values[1]:
        return 'B'
    elif price <= values[2]:
        return 'C'
    else:
        return 'D'
```

```python
dataset['category'] = dataset['price'].apply(categorize)
```

```python
pd.value_counts(dataset['category'])
```

```python
values
```

```python
pd.cut(dataset['price'], bins=values, labels=['A', 'B'])
```

```python
dataset.drop(['category'], axis=1)
```

```python
dataset.drop(dataset[dataset['bedrooms'] > 2].index, axis=0)
```

```python
dataset[dataset['bedrooms'] > 2].index
```
