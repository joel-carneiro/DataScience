## Criando novas variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv("extras/dados/aluguel.csv", sep = ";")
dados.head()
```

```python
dados['Valor bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head()
```

```python
dados['Valor m2'] = dados['Valor bruto'] / dados['Area']
```

```python
dados.head()
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
dados
```

```python
dados.drop(columns = ['Tipo Agregado'], inplace = True)
```

```python
dados.head()
```

```python
casa = ['Casa', 'Casa de Vila', 'Casa de Condomínio']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```
