```python
import pandas as pd
```

```python
dados = pd.read_csv('extras/dados/aluguel.csv', sep = ';')
```

```python
dados.head()
```

```python
dados['Bairro'].unique()
```

```python
bairros = ['Centro', 'Higienópolis', 'Cachambi', 'Grajaú',
       'Lins de Vasconcelos', 'Copacabana', 'Freguesia (Jacarepaguá)',
       'Barra da Tijuca', 'Tijuca', 'Olaria', 'Ipanema', 'Botafogo',
       'Recreio dos Bandeirantes', 'Leblon', 'Jardim Oceânico', 'Humaitá',
       'Méier', 'Vargem Pequena', 'Maracanã', 'Jacarepaguá',
       'Jardim Botânico', 'Vila Valqueire', 'Gávea', 'Cosme Velho',
       'Bonsucesso', 'Todos os Santos', 'Laranjeiras', 'Itanhangá',
       'Catete', 'Jardim Carioca', 'Praça Seca', 'Vila Isabel', 'Lagoa',
       'Taquara', 'Engenho Novo', 'Campo Grande', 'Flamengo']
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
```

```python
dados
```

```python
grupo_bairro = dados.groupby(['Bairro'])
```

```python
for bairro, dados in grupo_bairro:
    print('{} : R$ {}'.format(bairro, dados['Valor'].mean().round(2)))
```

```python
grupo_bairro['Valor'].describe()
```

```python
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20, 10))
```

```python
figure_max = grupo_bairro['Valor'].max().plot
figure_mean = grupo_bairro['Valor'].mean().plot
figure_min = grupo_bairro['Valor'].min().plot
```
