## Nova variável, aumento

```python
alucar.head()
```

```python
alucar['aumento'] = alucar['vendas'].diff()
```

```python
alucar.head()
```

```python
ax = sns.lineplot(x = 'mes', y = 'aumento', data = alucar)
```

```python
ax.figure.set_size_inches(12, 6)
ax.set_title('Aumento de vendas entre meses', loc = 'left', fontsize = 18)
ax.set_xlabel('Tempo', fontsize = 14)
ax.set_ylabel('Aumento entre meses (R$)', fontsize = 14)
ax.figure
```

```python
def criar_grafico(grafico, titulo, xlabel, ylabel, tamanhox, tamanhoy):
    grafico.set_title(titulo, loc = 'left', fontsize = 18)
    grafico.set_xlabel(xlabel, fontsize = 14)
    grafico.set_ylabel(ylabel, fontsize = 14)
    grafico.figure.set_size_inches(tamanhox, tamanhoy)
    return grafico
```

```python
grafico = sns.lineplot(x = 'mes', y = 'aumento', data = alucar)
```

```python
grafico = criar_grafico(grafico, 'Aumento x Tempo', 'Tempo', 'Aumento (R$)', 12, 6)
```

```python
grafico.figure
```
