```python
import pandas as pd
```

# Importando dataset

```python
vendas = pd.read_csv('data/chocolura.csv')
vendas.head()
```

```python
vendas['mes'] = pd.to_datetime(vendas['mes'])
```

```python
vendas.dtypes
```

```python
print('Quantidade de linhas e colunas: ', vendas.shape)
print('Quantidade de dados nulos: ', vendas.isna().sum().sum())
```

```python
vendas['aumento'] = vendas['vendas'].diff()
vendas['aceleracao'] = vendas['aumento'].diff()
```

```python
vendas.head()
```

# Plotando gr√°ficos

```python
import seaborn as sns
sns.set_style('darkgrid')
```

```python
grafico1 = sns.lineplot(x = 'mes', y = 'vendas', data = vendas)
grafico1.figure.set_size_inches(18, 8)
```

```python
grafico2 = sns.lineplot(x = 'mes', y = 'aumento', data = vendas)
grafico2.figure.set_size_inches(18, 8)
```

```python
grafico3 = sns.lineplot(x = 'mes', y = 'aceleracao', data = vendas)
grafico3.figure.set_size_inches(18, 8)
```

# Vendas di√°rias

```python
vendas_por_dia = pd.read_csv('data/vendas_por_dia.csv')
vendas_por_dia.head()
```

```python
print('Quantidade de linhas e colunas: ', vendas_por_dia.shape)
print('Quantidade de dados nulos: ', vendas_por_dia.isna().sum().sum())
```

```python
vendas_por_dia.dtypes
```

```python
vendas_por_dia['dia'] = pd.to_datetime(vendas_por_dia['dia'])
```

```python
vendas_por_dia.dtypes
```

```python
vendas_por_dia['aumento'] = vendas_por_dia['vendas'].diff()
vendas_por_dia['aceleracao'] = vendas_por_dia['aumento'].diff()
```

```python
vendas_por_dia
```

```python
grafico1 = sns.lineplot(x = 'dia', y = 'vendas', data = vendas_por_dia)
grafico1.figure.set_size_inches(18, 8)
```

```python
grafico2 = sns.lineplot(x = 'dia', y = 'aumento', data = vendas_por_dia)
grafico2.figure.set_size_inches(18, 8)
```

```python
grafico3 = sns.lineplot(x = 'dia', y = 'aceleracao', data = vendas_por_dia)
grafico3.figure.set_size_inches(18, 8)
```

```python
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia'].dt.day_name()
```

```python
vendas_por_dia
```

```python
vendas_por_dia['dia_da_semana'].unique()
```

```python
dias_traduzidos = {
    'Monday' : 'Segunda',
    'Tuesday': 'Ter√ßa',
    'Wednesday' : 'Quarta',
    'Thursday' : 'Quinta',
    'Friday' : 'Sexta',
    'Saturday' : 'S√°bado',
    'Sunday' : 'Domingo'
}
```

```python
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia_da_semana'].map(dias_traduzidos)
```

```python
vendas_por_dia
```

```python
vendas_agrupadas = vendas_por_dia.groupby('dia_da_semana')['vendas', 'aumento', 'aceleracao'].mean().round(2)
```

```python
vendas_agrupadas
```

```python
def plotar(x, y, data):
    grafico = sns.lineplot(x = x, y = y, data = data)
    grafico.figure.set_size_inches(18, 8)
    
    return grafico
```

```python
grafico5 = plotar('dia', 'vendas', vendas_por_dia)
```
