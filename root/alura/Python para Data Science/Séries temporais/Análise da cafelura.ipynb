```python
import pandas as pd
import seaborn as sns
```

```python
def plotar(x, y, data):
    sns.set_palette('Spectral')
    sns.set_style('whitegrid')
    grafico = sns.lineplot(x = x, y = y, data = data)
    grafico.figure.set_size_inches(10, 5)
    
    return grafico
```

```python
cafelura = pd.read_csv('data/cafelura.csv')
```

```python
cafelura.head()
```

```python
cafelura['mes'] = pd.to_datetime(cafelura['mes'])
```

```python
cafelura.dtypes
```

```python
cafelura['aumento'] = cafelura['vendas'].diff()
cafelura['aceleracao'] = cafelura['aumento'].diff()
```

```python
grafico1 = plotar(x = 'mes', y = 'vendas', data = cafelura)
```

```python
grafico2 = plotar(x = 'mes', y = 'aumento', data = cafelura)
```

```python
plotar(x = 'mes', y = 'aceleracao', data = cafelura)
```

```python
final_de_semana = pd.read_csv('data/dias_final_de_semana.csv')
final_de_semana.head()
```

```python
cafelura['vendas_normalizadas'] = cafelura['vendas'] / final_de_semana['quantidade_de_dias'].values
```

```python
cafelura.head()
```

```python
grafico4 = plotar(x = 'mes', y = 'vendas_normalizadas', data = cafelura)
```
