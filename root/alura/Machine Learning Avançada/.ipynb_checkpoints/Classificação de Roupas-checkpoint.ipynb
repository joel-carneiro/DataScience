> Com classificação de roupas

```python
# Importando bibliotecas necessárias
from tensorflow import keras
from matplotlib import pyplot as plt
```

```python
# Setando o dataset com os dados
df = keras.datasets.fashion_mnist

# Divindindo entre treino e teste
(imagens_treino, identificacoes_treino), (imagens_teste, identificacoes_teste) = df.load_data()
```

```python
# Explorando dados
print('-=-' * 15)

print('Dimensões do treino (input):', imagens_treino.shape)
print('Dimensões do teste (input): ', imagens_teste.shape)
print('Identificações do treino (output):', identificacoes_treino.shape[0])
print('Identificações do teste (output):', identificacoes_teste.shape[0])

print('-=-' * 15)
```

```python
# Renomeando as classificações
nomes_das_classificacoes = [
    'Camiseta', 'Calça', 'Pullover', 'Vestido', 'Casaco', 'Sandália', 'Camisa', 'Tênis', 'Bolsa', 'Bota'
]

# Loop para plotar todas as imagens
for imagem in range(10):
    # Definindo a figura
    plt.subplot(2, 5, imagem + 1)
    
    # Mostrando a imagem
    plt.imshow(imagens_treino[imagem])
    
    # Definindo o título das imagens
    plt.title(nomes_das_classificacoes[identificacoes_treino[imagem]])
```
