```python
import pandas as pd
```

```python
filmes = pd.read_csv('/home/jdspy/Documentos/datasets/movies.csv')
filmes.columns = ['filme_id', 'titulo', 'generos']
filmes.index = filmes['filme_id']
filmes.head()
```

```python
notas = pd.read_csv('/home/jdspy/Documentos/datasets/ratings.csv')
notas.columns = ['usuario_id', 'filme_id', 'notas', 'momento']
notas.head()
```

```python
notas.describe()
```

```python
total_de_votos = pd.value_counts(notas['filme_id']); total_de_votos
```

```python
filmes.loc[356]
```

```python
filmes['total_de_votos'] = total_de_votos
```

```python
filmes.head()
```

```python
filmes.sort_values(by='total_de_votos', ascending=False)
```

```python
notas_medias = notas.groupby('filme_id')['notas'].mean()
```

```python
filmes['notas_medias'] = notas_medias
```

```python
filmes.sort_values(by='total_de_votos', ascending=False)
```

```python
filmes.query('total_de_votos > 50').sort_values(by='notas_medias', ascending=False)
```
