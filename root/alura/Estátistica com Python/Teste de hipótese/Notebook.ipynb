## Teste de normalidade

```python
# Configurando o jupyter para não completar automaticamente
%config IPCompleter.greedy=False
```

```python
# Importando dataset do projeto
import pandas as pd
df = pd.read_csv('dados.csv')
df.head()
```

```python
# Histograma da variável renda
df['Renda'].hist(bins = 50, alpha = 0.35)
```

```python
# Função para executar o teste de normalidade
from scipy.stats import normaltest
```

```python
# Caso pvalue < 0.05, rejeitar H0
statistic, pvalue = normaltest(df['Renda'])
print(f'pvalue = {pvalue}')
print(f'statistic = {statistic}')
```

```python
# Histograma da variável altura
df['Altura'].hist(bins = 50, alpha = 0.35)
```

```python
# Caso pvalue < 0.05, rejeitar H0
statistic, pvalue = normaltest(df['Altura'])
print(f'pvalue = {pvalue}')
print(f'statistic = {statistic}')
```

```python
# Função para descobrir se a Series é uma distribuição normal
def is_normal(s):
    statistic, pvalue = normaltest(s)
    
    return pvalue > 0.05
```

```python
print('Altura é uma variável com distribuição normal:', is_normal(df['Altura']))
print('Renda é uma variável com distribuição normal:', is_normal(df['Renda']))
```
