# Introdução a Estatística

Estatística é um conjunto de métodos usados para se analisar dados. A Estatística pode ser aplicada em praticamente todas as áreas do conhecimento humano e em algumas áreas recebe um nome especial. Este é o caso da Bioestatística, que trata de aplicações da Estatística em Ciências Biológicas e da Saúde.

```python
import pandas as pd
```

```python
dataset = pd.read_csv('data.csv')
```

# 2 - Representações Gráficas

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Histograma
sns.histplot(dataset['Altura'])

plt.title('Histograma')
```

```python
# Suavização
sns.kdeplot(dataset['Altura'])

plt.title('Densidade de Frequências')
```

```python
sns.distplot(dataset['Altura'])
```

```python
frequency = dataset['Altura'].value_counts().sort_index()
accumulative_frequency = frequency.cumsum()
```

```python
accumulative_frequency
```

```python
accumulative_frequency.plot()

plt.title('Probabilidade acumulada')
```

```python
sns.countplot(x=dataset['Cor'])
```

```python
sns.countplot(y=dataset['UF'].sort_index())
```
