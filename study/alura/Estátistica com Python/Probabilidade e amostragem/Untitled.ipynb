```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('data/dados.csv')
df = df.reset_index()
```

```python
df.columns = ['NUMERO', 'DDD', 'SEXO', 'IDADE', 'COR', 'ESTUDO', 'RENDA', 'ALTURA']
```

```python
df
```

```python
sexo = {
    0: 'Masculino',
    1: 'Feminino'
}

df['SEXO'] = df['SEXO'].replace(sexo)
```

```python
df['COR'].unique()
```

```python
cor = {
    6 : 'Pardo',
    0 : 'Não informado',
    4 : 'Amarelo',
    2 : 'Negro',
    8 : 'Branco'
}

df['COR'] = df['COR'].replace(cor)
```

```python
df['ALTURA'] = df['ALTURA'].apply(lambda x : round(x, 2))
```

```python
df
```

```python
# Criação da conexão com o banco de dados
engine = create_engine('mysql+pymysql://root:thisissimple@localhost:3306/POPULACAO')

# Escrevendo o DataFrame em uma tabela do MySQL
df.to_sql('TABELA', engine, if_exists='replace')
```
