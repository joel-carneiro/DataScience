```python
# Importando o módulo pandas para análise de dados
import pandas as pd

# Definindo URI's para extração de dados
uri = {
    'confirmed' : 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv',
    'deaths' : 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv',
    'recovered' : 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
}

# Casos confirmados
confirmed = pd.read_csv(uri['confirmed'])

# Mortes confirmadas
deaths = pd.read_csv(uri['deaths'])

# Recuperações confirmadas
recovered = pd.read_csv(uri['recovered'])
```

```python
# Visualizando os 5 primeiros dados de cada dataset
display(confirmed.head())
display(deaths.head())
display(recovered.head())
```

```python
# Categorizando observações caso seja china ou não
is_china = confirmed['Country/Region'] == 'China'; is_china

confirmed['is_china'] = is_china
```

```python
course_date = '3/2/20'
```

```python
# Plotando o gráfico de pizza
confirmed.groupby(by = 'is_china').sum()[course_date].plot(kind = 'pie')
```

```python
# Plotando o gráfico de barra
confirmed.groupby(by = 'is_china').sum()[course_date].plot(kind = 'bar')
```

```python
import numpy as np

confirmed['is_china'] = np.where(confirmed['is_china'], 'China', 'Others')
```

```python
confirmed.groupby(by = 'is_china').sum()['3/2/20'].plot(kind = 'bar')
```

```python
summed = confirmed.groupby(by = 'is_china').sum()
summed.T[2:47].plot()
```

```python
summed.T[2:47].diff().plot()
```
