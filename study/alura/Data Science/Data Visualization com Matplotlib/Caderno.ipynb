```python
import datetime

fig = plt.figure(figsize = (12, 6))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(tempo['data'], tempo['temperatura'], color = 'teal')

eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))

eixo.set_title('Temperatura no momento', fontsize = 25)
eixo.set_ylabel('Temperatura', fontsize = 20)
eixo.set_xlabel('Data', fontsize = 20)
eixo.legend(['Temperatura'], loc = 'lower right', fontsize = 15)
```

```python
import datetime

fig = plt.figure(figsize = (12, 6))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(tempo['data'], tempo['temperatura'], color = 'teal', marker = 'o')

eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))

eixo.set_title('Temperatura no momento', fontsize = 25)
eixo.set_ylabel('Temperatura', fontsize = 20)
eixo.set_xlabel('Data', fontsize = 20)
eixo.legend(['Temperatura'], loc = 'lower right', fontsize = 15)
```

```python
import datetime

fig = plt.figure(figsize = (12, 6))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(tempo['data'], tempo['temperatura'], color = 'teal')

eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))

eixo.set_title('Temperatura no momento', fontsize = 25)
eixo.set_ylabel('Temperatura', fontsize = 20)
eixo.set_xlabel('Data', fontsize = 20)
eixo.legend(['Temperatura'], loc = 'lower right', fontsize = 15)

eixo.grid(True)
```

```python
fig = plt.figure(figsize = (15, 8))
eixo1 = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.6, 0.6, 0.4, 0.4])

eixo1.plot(tempo['data'], tempo['temperatura'], color = 'orange')
eixo1.set_xlim(datetime.datetime(2014, 5, 1), datetime.datetime(2014, 6, 1))
eixo1.grid(True)
eixo1.set_title('Temperatura no mês de maio em 2014', fontsize = 25)
eixo1.set_xlabel('Momento', fontsize = 20)
eixo1.set_ylabel('Temperatura', fontsize = 20)
eixo1.legend(['Temperatura'], loc = 'upper left', fontsize = 15)
eixo1.set_ylim(270, 320)

eixo2.plot(tempo['data'], tempo['temperatura'], color = 'teal')
eixo2.set_xlim(datetime.datetime(2014, 1, 1), datetime.datetime(2015, 1, 1))
eixo2.grid(True)
```

```python
import datetime

fig = plt.figure(figsize = (12, 6))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(tempo['data'], tempo['temperatura'], color = 'teal')

eixo.set_title('Temperatura no momento', fontsize = 25)
eixo.set_ylabel('Temperatura', fontsize = 20)
eixo.set_xlabel('Data', fontsize = 20)
eixo.legend(['Temperatura'], loc = 'lower right', fontsize = 15)

eixo.axhline(max(tempo['temperatura']), color = 'k', linestyle = '--')
eixo.axhline(min(tempo['temperatura']), color = 'k', linestyle = '--')
```
