## Sobre


O ouro sempre ocupou um lugar de destaque na economia do país e entre as pessoas. Neste notebook, irei tentar prever o preço do ouro usando algoritmos de aprendizado de máquina

Os preços do ouro dispararam historicamente, chegando a US$ 1.800 em 2020. Essas oscilações que o ouro enfrentou podem ser atribuídas a vários eventos ocorridos durante o período. Os preços das ações estão fortemente correlacionados com informações públicas e eventos mundiais, e o ouro não é exceção.

Ao contrário de outros ativos financeiros, os metais preciosos se comportam positivamente em momentos de crise. De fato, diante de eventos como epidemias, guerras e perturbações econômicas, a maioria dos ativos sofre uma queda em seu valor. No entanto, em busca de cobertura, os investidores estão voltando sua atenção para o ouro. Como resultado, a demanda por metal amarelo aumenta, resultando em um aumento na valorização do mercado.

O ouro tende a ser visto como um ativo de investimento estratégico e tático ao considerar a diversificação do portfólio. No entanto, os aspectos econômicos e financeiros não são os únicos fatores que podem afetar os preços do ouro. Fatores geopolíticos também podem influenciar a valorização do mercado do metal amarelo.

De fato, mostram que diante de riscos geopolíticos, o ouro atua como um porto seguro. De fato, ele tem uma reação afirmativa em momentos de ameaças decorrentes de tensões geopolíticas e internacionais. Ao analisar dados financeiros, geopolíticos e globais sobre desastres naturais, é possível prever as flutuações futuras do preço do ouro, o que permitirá aos investidores melhorar a gestão de sua carteira diante de movimentos inesperados. do mercado.

## 1.0 Ferramentas necessárias

```python
# Módulos necessários para manipulação de dados
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
# Desativando possíveis avisos
warnings.filterwarnings('ignore')
```

## 2.0 Coleta de dados

```python
# Coletando dados na nuvem
df = pd.read_csv('https://raw.githubusercontent.com/amankharwal/Website-data/master/gold_price.csv')
```

```python
# Visualizando as 5 primeiras observações
df.head()
```

## 3.0 Limpeza de dados

```python
# Visualizando as 20 primeiras observações
df.head(20)
```

```python
# Visualizando informações
df.info()
```

```python
# Tratando dtype da variável "Date"
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Visualizando dados nulos
null = df.isna().sum().sum()
```

```python
df.shape
```

```python
print('Porcentagem de dados nulos: %.2f%%' % ((null / df.shape[0]) * 100))
```

```python
# Considerando os dados nulos irrelevantes, irei remover
df.dropna(inplace=True)
```

## 4.0 Exploração de dados

```python
plt.style.use('ggplot')

plt.figure(figsize=(10, 5))

plt.plot(df['Date'], df['USD (AM)'])
plt.title('Série temporal do preço do ouro em Dolar')
plt.xlabel('Data')
plt.ylabel('Preço ($)')

plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.plot(df['Date'], df['EURO (AM)'])
plt.title('Série temporal do preço do ouro em Euro')
plt.xlabel('Data')
plt.ylabel('Preço ($)')

plt.show()
```

## 5.0 Modelando algoritmos de aprendizagem de máquina

### 5.1 Separando dados de treino e teste

```python
from sklearn.model_selection import train_test_split

x = df[['EURO (AM)']]
y = df['USD (AM)']

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=101, test_size=0.3)
```

### 5.2 Modelando um algoritmo de regressão linear

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train, y_train)
```

```python
# Acurácia usando dados de output como dolar e input como euro
model.score(x_test, y_test)
```

```python
plt.scatter(x_test, y_test)

plt.scatter(model.predict(x_test), y_test)

plt.legend(['Real', 'Predito'])

plt.show()
```

### 5.3 Modelando em relação a data

```python
df['Year'] = df['Date'].apply(lambda x : x.year)
df['Month'] = df['Date'].apply(lambda x : x.month)
df['Day'] = df['Date'].apply(lambda x : x.day)
```

```python
df['USD'] = (df['USD (AM)'] + df['USD (PM)']) / 2
```

```python
x = df[['Year', 'Month', 'Day']]
y = df['USD']

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=101, test_size=0.3)
```

```python
model = LinearRegression()

model.fit(x_train, y_train)

model.score(x_test, y_test)
```

```python
plt.plot(df['Year'], model.predict(x))
plt.plot(df['Year'], df['USD'])
```
