```python
remove_vars = null_data[null_data > len(raw_data) * 0.5]
```

```python
raw_data = raw_data.drop(remove_vars.index, axis=1)
```

```python
null_data = raw_data.isna().sum() / len(raw_data); null_data
```

```python
raw_data.classificacaoFinal.dropna()
```

```python
raw_data.columns
```

```python
dataset = raw_data[['sintomas', 'racaCor', 'sexo', 'classificacaoFinal']]
```

```python
dataset.info()
```

```python
dataset = dataset.dropna()
```

```python
dataset
```

```python
lista_de_sintomas = set()

for ob in dataset.sintomas:
    for item in ob.split(','):
        item = item.strip()
        lista_de_sintomas.add(item)
```

```python
lista_de_sintomas
```

```python
lista_de_sintomas = list(lista_de_sintomas)
```

```python
lista_de_sintomas
```

```python
for i in range(len(lista_de_sintomas)):
    dataset[lista_de_sintomas[i]] = dataset.sintomas.apply(lambda x : lista_de_sintomas[i] in x)
```

```python
dataset = dataset.replace({True: 1, False: 0})
```

```python
dataset
```

```python
dataset.sexo.unique()
```

```python
dataset = dataset.drop(dataset.query('sexo == "Ceará"').index[0])
```

```python
dataset.sexo.unique()
```

```python
dataset.classificacaoFinal.unique()
```

```python
dataset
```

```python
dataset.classificacaoFinal.value_counts()
```

```python
dataset.classificacaoFinal = dataset.classificacaoFinal.replace({
    "Descartado" : "NEGATIVO",
    "Confirmado Laboratorial" : "POSITIVO",
    "Síndrome Gripal Não Especificada" : "POSITIVO NÃO-ESPECIFICADO",
    "Confirmado Clínico-Epidemiológico" : "POSITIVO",
    "Confirmado por Critério Clínico" : "POSITIVO",
    "Confirmado Clínico-Imagem" : "POSITIVO"
})
```

```python
dataset
```

```python
input_data = dataset.iloc[:, 4:]
```

```python
output_data = dataset.classificacaoFinal
```

```python
dataset = pd.concat([input_data, output_data], axis=1)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset = dataset.rename(columns={'classificacaoFinal' : 'Resultado'})
```
