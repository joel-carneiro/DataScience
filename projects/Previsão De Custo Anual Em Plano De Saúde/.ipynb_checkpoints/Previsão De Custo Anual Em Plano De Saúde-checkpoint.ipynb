## 2.2 - Análisando a Variável Alvo

```python
# Função útil para visualização de distribuição de frequência
def view_frequency_distribution(array, title):
    plt.figure(figsize=(10, 5))
    sns.histplot(array, kde=True)
    plt.title(title + ' (Histograma)')
    plt.show()
    
    plt.figure(figsize=(10, 5))
    sns.boxplot(array)
    plt.title(title + ' (Boxplot)')
    plt.show()

    
def count_categorical_variables(array, title):
    sns.countplot(array, palette='Set2')
    plt.title(title)
    plt.show()
```

```python
# Visualizando as despesas com plano de saúde em distribuição de frequência
view_frequency_distribution(train_data['charges'], title='Despesas com plano de saúde')
```

# 3 - Pré-processamento de Dados

```python
# Arredondando a variável Idade
train_data['age'] = round(train_data['age'])
```

```python
# Transformando variáveis categoricas em númericas
train_data = pd.get_dummies(train_data, drop_first=True)
```

```python
# Visualizando nova aparência da base de dados
train_data.head(3)
```

```python
# Separando entre entrada e saída
x = train_data.drop(['charges'], axis=1)
y = train_data['charges']
```

```python
x.head(3)
```

```python
y.head(3)
```

```python
# Separando entre dados de treino e dados de teste
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, test_size=0.25)
```

# 4 - Construção do Modelo Preditivo

```python
# Instânciando o modelo de regressão linear
from sklearn.linear_model import LinearRegression

linear = LinearRegression()
```

```python
# Instânciando o modelo de regressão ridge
from sklearn.linear_model import Ridge

ridge = Ridge()
```

```python
# Instânciando o modelo de regressão lasso
from sklearn.linear_model import Lasso

lasso = Lasso()
```

```python
# # Instânciando o modelo de regressão de floresta aleatória
from sklearn.ensemble import RandomForestRegressor

random_forest = RandomForestRegressor() 
```

```python
# Array com os modelos
models = [
    linear,
    ridge,
    lasso,
    random_forest
]
```

# 5 - Avaliando o Modelo Preditivo

```python
from sklearn import metrics
```

```python
# Treinando e avaliando cada modelo com as métricas r2 e erro médio absoluto
for model in models:
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    
    r2_score = metrics.r2_score(y_test, y_pred)
    mean_absolute_error = metrics.mean_absolute_error(y_test, y_pred)

    print('MODELO <=> %s' % model.__str__()[:-2])
    print('\t r2: %.4f' % r2_score)
    print('\t MAE: %.4f' % mean_absolute_error)
    print('\n')
```

# 6 - Melhorando o Modelo

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Achando os melhores parâmetros
model = RandomForestRegressor()

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [2, 3, 4, 5, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=1, verbose=2)
grid_search.fit(x, y)
```

```python
# Melhor modelo possui esses parâmetros
grid_search.best_params_
```

```python
# Criando o melhor modelo de árvore aleatória
model = RandomForestRegressor(**grid_search.best_params_)
```

```python
# Treinando o modelo 
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
# Métrica R2
metrics.r2_score(y_test, y_pred)
```

```python
# Métrica erro médio absoluto
metrics.mean_absolute_error(y_test, y_pred)
```
