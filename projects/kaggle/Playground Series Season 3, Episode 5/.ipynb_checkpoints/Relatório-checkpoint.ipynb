## 1.2 Desbalanceamento de Classe

Até agora, estamos lidando com múltiplas classes ao inves de um simples modelo binário, oque pode indicar um tendência de regressão, mas veremos isso mais adiante, mas por enquanto, vamos supor que temos classes

```python
sns.set_style('darkgrid')
sns.set_palette('Set2')

_ = sns.countplot(x='quality', data=train, alpha=0.8)
_.set_title('Balanceamento de Classes')
_.set_xlabel('Qualidade')
_.set_ylabel('Frequência')

_.figure.savefig('balanceamento_de_classes.png')
```

**Observações sobre o balanceamento de classe**

- A maioria das classifações é 5 ou 6, oque significa que apenas usar 5 como métrica de qualidade obterá uma "kappa metric" razoável a partir dele

- Temos pouquissímas intâncias de classe 3, 4 ou 8
    - Esses casos raros serão dificeis de pontuar corretamente

- A classificação não assume valores acima de 8, e também não há valores abaixo de 3
    - Portando todas as classificações devem estar entre 3 e 8

- Apesar do problema parecer com o de classificação, talvez seja útil tratá-lo como de regressão e arredondar o valor para a classificação mais próxima

## 1.3 - Valores nulos

Agora, vamos explorar os dados faltantes, para evitar problemas sistemáticos com o modelo

```python
labels = ['Não Nulos', 'Nulos']
values = [train.shape[0], train.isna().sum().sum()]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

axes[0].pie(values, labels=labels)

labels = ['Não Nulos', 'Nulos']
values = [train.shape[0], train.isna().sum().sum()]

axes[1].pie(values, labels=labels)

axes[0].set_title('Dados nulo em treino')
axes[1].set_title('Dados nulo em teste')

plt.show()
```
