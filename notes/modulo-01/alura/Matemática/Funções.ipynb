```python
# Importando módulo para computação símbolica
import sympy as sp
from sympy import plot as splot
```

```python
# Instânciando uma variável representada por "x"
x = sp.Symbol('x')

# Descrevendo uma função
f = x ** 2 + 1

# Calculando o limite da função quando x tende a 2
lim = sp.limit(f, x, 2)

# Exibindo valores
print('f(x) = ', f)
print('lim x-> 2 f(x) = ', lim)
```

```python
# Instânciando funções
f = x ** 2
g = x ** 3

# Propriedades dos limites I
print('Propriedade I:', sp.limit(f + g, x, 1) == sp.limit(f, x, 1) + sp.limit(g, x, 1))

# Propriedades dos limites II
print('Propriedade II', sp.limit(f / g, x, 1) == sp.limit(f, x, 1) / sp.limit(g, x, 1))

# Propriedades dos limites III
print('Propriedade III', sp.limit(f * g, x, 1) == sp.limit(f, x, 1) * sp.limit(g, x, 1))

# Propriedades dos limites IV
print('Propriedade IV', sp.limit((f * 5), x, 1) == sp.limit(f, x, 1) * 5)
```

```python
# Função para calculo de variação média de uma função
def average_variation(f, x, a, b):
    
    # Calculando a diferença entre f(x)1 e f(x)2
    top = sp.limit(f, x, b) - sp.limit(f, x, a)
    
    # Calculando a distância entre os limites 
    bot = b - a
    
    # Retornando a variação
    return top / bot

print('Variação média de f(x) no intervalo [1, 10]:', average_variation(f, x, 1, 10))
```

```python
# Módulo para manipulação de dados
import pandas as pd

# Definindo função de queda livre
def free_fall_analysis(h):
    # Definindo constantes
    G = 9.8
    T = 0
    h_list = []
    
    # Calculo da distância 
    while True:
        h -= G * (T ** 2) / 2
        if h < 0:
            break
        h_list.append(h)
    return pd.Series(h_list)
        
        
free_fall_analysis(100)
```
