## Importando bibliotecas necessárias

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

## Criando o dataframe

```python
ensaios = np.array([[-1, -1], [-1, 1], [1, -1], [1, 1]])

experimentos = pd.DataFrame(ensaios, columns=['Chocolate', 'Farinha'])
experimentos
```

```python
experimentos['Porcoes'] = [19, 37, 24, 49]
```

```python
experimentos
```

## Configurando o Seaborn

```python
sns.set_palette('terrain')
sns.set_style('darkgrid')
```

## Analisando graficamente os resultados

```python
ax1 = sns.lmplot(x = 'Farinha', y = 'Porcoes', data = experimentos, ci=None, hue = 'Chocolate')
```

```python
sns.lmplot(x = 'Chocolate', y = 'Porcoes', data = experimentos, ci = None, hue = 'Farinha')
```
