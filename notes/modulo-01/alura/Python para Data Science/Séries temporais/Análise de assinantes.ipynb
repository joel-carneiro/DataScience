```python
import pandas as pd
```

# Importando dataset e configurações iniciais

```python
assinantes = pd.read_csv('data/newsletter_alucar.csv')
assinantes.head()
```

```python
assinantes['mes'] = pd.to_datetime(assinantes['mes'])
```

```python
print('Quantidade de linhas e colunas: ', assinantes.shape)
print('Quantidade de dados nulos: ', assinantes.isna().sum().sum())
```

```python
assinantes['aumento'] = assinantes.assinantes.diff()
```

```python
assinantes['aceleracao'] = assinantes.aumento.diff()
```

```python
assinantes.head()
```

# Graficos de assinantes, aumento e aceleração

```python
import seaborn as sns
sns.set_style('darkgrid')
```

```python
grafico1 = sns.lineplot(x = 'mes', y = 'assinantes', data = assinantes)
grafico1.figure.set_size_inches(12, 6)
```

```python
grafico2 = sns.lineplot(x = 'mes', y = 'aumento', data = assinantes)
grafico2.figure.set_size_inches(12, 6)
```

```python
grafico3 = sns.lineplot(x = 'mes', y = 'aceleracao', data = assinantes)
grafico3.figure.set_size_inches(12, 6)
```
