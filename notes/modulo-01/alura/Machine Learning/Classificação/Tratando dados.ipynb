```python
import pandas as pd
```

```python
dados = pd.read_csv('customer.csv')
dados.head()
```

```python
dados.shape
```

```python
traducao = {'Sim' : 1, 'Nao' : 0}

dadosmodificados = dados[['Conjuge', 'Dependentes', 'TelefoneFixo', 'PagamentoOnline', 'Churn']].replace(traducao)
dadosmodificados
```

```python
dummies_dados = pd.get_dummies(dados.drop(['Conjuge', 'Dependentes', 'TelefoneFixo', 'PagamentoOnline', 'Churn'], axis = 1))
dummies_dados
```

```python
dados_final = pd.concat([dadosmodificados, dummies_dados], axis = 1)
dados_final = dados_final.drop(['Unnamed: 0'], axis = 1)

dados_final.head()
```

```python
dados_final.to_csv('customer-tratado.csv')
```
