# Visualização de dados

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
dataset = pd.read_csv('/home/jdspy/Documentos/datasets/monthly_csv.csv')
```

```python
dataset['Date'] = dataset['Date'].astype('datetime64')
```

```python
plt.style.use('seaborn')

plt.figure(figsize=(10, 5))

plt.title('Preço do ouro entre 1950 e 2023')

plt.xlabel('Data')

plt.ylabel('Preço (U$)')

plt.plot(dataset['Date'], dataset['Price'])
```
