```r
library(ggplot2)
```

```r
x <- rnorm(n=1000, mean=100, sd=20)
y <- x + sample(1:35, size=length(x), replace=TRUE)
z <- sample(c('RJ', 'SP', 'MG', 'ES'), size=length(x), replace=TRUE)

df <- data.frame(size=x, price=y, local=z)

head(df, 3)
```

```r
ggplot(df, aes(x=size, y=price)) +
    geom_point(shape=10, color='darkgreen') +
    geom_smooth(method = lm, color = 'darkblue', se = FALSE)
```
