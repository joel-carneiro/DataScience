```r
randint <- function (min, max, size) {
    n <- sample(min:max, size=size, replace=TRUE)
    return (n)
}
```

```r
# Com a par(), podemos definir os parâmetros visuais do nosso gŕafico
par(pch='D')
```

```r
x <- 1:100
y <- x + randint(1, 25, 100)

plot(x, y)
par(pch='1')
plot(x, y)
```

```r
par(mfrow=c(2, 2), col.axis='blue')

plot(x, y, main='LAS 0')
plot(y, x, main='LAS 1')
plot(x^-1, y, main='LAS 2')
plot(x, y^-1, main='LAS 3')
```

```r
dist_normal_1 <- rnorm(n=100, mean=175, sd=10)

dist_poisson_1 <- rpois(n=100, lambda=175)

dist_normal_2 <- rnorm(n=100, mean=150, sd=30)

dist_poisson_2 <- rpois(n=100, lambda=150)
```

```r
hist(dist_normal_1)
```

```r
hist(dist_poisson_1)
```

```r
# Define o tamanho da figura padrão
options(repr.plot.width=8, repr.plot.height=8)

plot(dist_normal_1, dist_poisson_1, col='darkgreen', pch=10)

points(dist_normal_2, dist_poisson_2, col='darkblue', pch=10)

points(dist_normal_1, dist_normal_2, col='darkred', pch=10)

points(dist_poisson_2, dist_poisson_1, col='darkgrey', pch=10)
```
