```r
library('ggplot2')
```

```r
x <- 10:1000
y <- x + sample(10:100, size=length(x), replace=TRUE)
z <- sample(c('SP', 'RJ'), size=length(x), replace=TRUE)

df <- data.frame(size=x, price=y, local=z)

head(df, 3)
```

```r
qplot(df$size, df$price)
```

```r
camada1 <- geom_point(
    mapping = aes(x=df$size, y=df$price, color=df$local),
    size=3
)

ggplot() + camada1
```

```r
ggplot(df, aes(x=size, y=price)) +
    geom_point(aes(color=local)) +
    geom_smooth(method = 'lm')
```
