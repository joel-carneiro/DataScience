## 1.1 - Objeto Series

```python
# O objeto series funciona como um array numpy, mas com algumas diferenças
meses = pd.Series(np.arange(1, 13, 1))
meses
```

```python
# O index das series podem ser referenciados por strings também, ao contrário de arrays
ms = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
meses.index = ms
meses
```

```python
# Podemos nos referenciar atráves do index modificado
meses['Mar']
```

```python
# Slicing atráves de strings também é possível
meses['Jun':'Dez']
```

```python
# Definindo um conjunto de dados aleatórios em formato de times-series
np.random.seed(100) # Definindo um número fixo para geração de números aleatórios
data = abs(np.random.randn(12)) * 1000
lucros = pd.Series(data=data, index=ms)
lucros
```

```python
# Agora podemos visualizar nossa time series (valor ao decorrer do tempo)
plt.plot(lucros)
```

```python
# Podemos calcular a média da nossa série temporal
lucros.mean()
```

```python
# Podemos calcular o desvio padrão da nossa série temporal
lucros.std()
```

```python
# Como também pode fazer um trabalho mais agilizado
lucros.describe()
```

```python
# Podemos ainda fabricar gráficos a partir de atributos do nosso objeto series
plt.bar(x = lucros.index, height = lucros.values)
```

```python
lucros_ordenados = lucros.sort_values(ascending=False)
plt.bar(x = lucros_ordenados.index, height = lucros_ordenados.values)
```

```python
lucros_ordenados = lucros.sort_values(ascending=True)
plt.bar(x = lucros_ordenados.index, height = lucros_ordenados.values)
```

```python
# Vejamos como se organiza o primeiro objeto series
lucros
```

```python
# Vejamos agora outro objeto series
corr = ms[5:]
np.random.seed(100) # Definindo um número fixo para geração de números aleatórios
data = abs(np.random.randn(7)) * 1000
corr = pd.Series(data=data, index=corr)
corr
```

```python
# Ao somarmos, chegaremos a valores NaN
lucros + corr
```

```python
# Reindexando a serie
s1 = lucros.reindex(ms, fill_value=0)
s1
```

```python
# Reindexando a serie
s2 = corr.reindex(ms, fill_value=0)
s2
```

```python
# Agora sim, somando valores sem gerar NaN
s1 + s2
```

```python
# Podemos aplicar uma função a cada elemento da series
def tribute(value):
    value = value * 0.15
    return value
```

```python
# Aplicando o imposto a cada mês
lucros.map(tribute)
```

```python
# Aplicando o imposto a cada mês
lucros.apply(tribute)
```
