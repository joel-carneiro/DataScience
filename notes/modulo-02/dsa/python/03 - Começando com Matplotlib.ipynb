```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
x = np.linspace(15, 300, 100)
y = x * 5 + np.random.normal(0, 150, 100)
```

```python
plt.scatter(x, y)
```

```python
alpha, beta, rvalue, pvalue, std = linregress(x, y)
```

```python
y_pred = x * alpha + beta
```
