# 2 - Trabalhando com Machine Learning

## 2.1 - Preparação de Dados

````python
# Visualizando o padrão dos dados
diabetes.head()
````

````python
# Função para padronização de dados
def min_max_scaler(x):
    return (x - x.min()) / (x.max() - x.min())
````

<font size=5> $$ x_s = \frac{x - x_\min}{x_\max - x_\min}$$ </font>

````python
# Alterando a escala dos dados com MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

x = diabetes.drop(columns=['class'])
y = diabetes['class']

scaler = MinMaxScaler()

x_s = scaler.fit_transform(x)

x_s
````

````python
# Função para padronização de dados
def standard_scaler(x):
    return (x - x.mean()) / x.std()
````

<font size=5> $$ x_s = \frac{x - \mu}{\sigma}$$ </font>

````python
# Alterando a escala dos dados com StandardScaler
from sklearn.preprocessing import StandardScaler

x = diabetes.drop(columns=['class'])
y = diabetes['class']

scaler = StandardScaler()

x_s = scaler.fit_transform(x)

x_s
````

<font size=5> $$ X_{\text{s}} = \frac{X}{||X||} $$

````python
# Alterando a escala dos dados com Normalizer
from sklearn.preprocessing import Normalizer

x = diabetes.drop(columns=['class'])
y = diabetes['class']

scaler = Normalizer()

x_s = scaler.fit_transform(x)

x_s
````

<font size=5>$$ X_{\text{s}} = \begin{cases} 
      0 & \text{se } X \leq \text{T} \\
      1 & \text{se } X > \text{T} 
   \end{cases} $$


````python
# Alterando a escala dos dados com Binarizer
from sklearn.preprocessing import Binarizer

x = diabetes.drop(columns=['class'])
y = diabetes['class']

scaler = Binarizer(threshold=0)

x_s = scaler.fit_transform(x)

x_s
````

## 2.2 - Seleção de Features

````python
# Importando um objeto para seleção das K melhores variáveis
from sklearn.feature_selection import SelectKBest

# Função que será usada para cálculo
from sklearn.feature_selection import chi2

# Separando dados entre dados de entrada e saída
x = diabetes.drop(columns=['class'])
y = diabetes['class']

# Instânciando o objeto com a função chi2, e escolhendo as 4 melhores features
selector = SelectKBest(score_func=chi2, k=4)

# Treinando o objeto com os dados
selector.fit(x, y)

# Selecionando as 4 melhores features
selector.get_feature_names_out()
````

````python
# Importando a função ExtraTreesClassifier
from sklearn.ensemble import ExtraTreesClassifier

# Separando dados entre dados de entrada e saída
x = diabetes.drop(columns=['class'])
y = diabetes['class']

# Criando e treinando o modelo
ext = ExtraTreesClassifier()
ext.fit(x, y)

# Visualizando cada score de cada variável
for feature, value in zip(ext.feature_names_in_, ext.feature_importances_):
    print(f'Feature: [{feature}] -> [{value:.2f}]')
````

## 2.3 - Validando Modelos

````python
# Importando função de regressão logística
from sklearn.linear_model import LogisticRegression

# Importando função para separar dados
from sklearn.model_selection import train_test_split

# Separando dados entre entrada e saída
x = diabetes.drop(columns=['class'])
y = diabetes['class']

# Separando dados entre dados de treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=100, test_size=0.25)

# Instânciando um objeto de regressão logística
model = LogisticRegression(solver='lbfgs', max_iter=1000)

# Treinando modelo
model.fit(x_train, y_train)

# Acurácia do modelo
model.score(x_test, y_test)
````

````python
# Importando função de regressão logística
from sklearn.dummy import DummyClassifier

# Importando função para separar dados
from sklearn.model_selection import train_test_split

# Separando dados entre entrada e saída
x = diabetes.drop(columns=['class'])
y = diabetes['class']

# Separando dados entre dados de treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=100, test_size=0.25)

# Instânciando um objeto de regressão logística
model = DummyClassifier()

# Treinando modelo
model.fit(x_train, y_train)

# Acurácia do modelo
model.score(x_test, y_test)
````

````python
# Importando função de regressão logística
from sklearn.linear_model import LogisticRegression

# Importando função para validação cruzada
from sklearn.model_selection import cross_validate

# Separando dados entre entrada e saída
x = diabetes.drop(columns=['class'])
y = diabetes['class']

# Instânciando um objeto de regressão logística
model = LogisticRegression(solver='lbfgs', max_iter=1000)

# Aplicando a validação de cruzada
result = cross_validate(model, x, y, cv = 10)

# Visualizando a acurácia média
result['test_score'].mean()
````

````python
# Importando função de regressão logística
from sklearn.linear_model import LogisticRegression

# Importando função para validação cruzada
from sklearn.model_selection import KFold, cross_validate

# Separando dados entre entrada e saída
x = diabetes.drop(columns=['class'])
y = diabetes['class']

# Instânciando um objeto de regressão logística
model = LogisticRegression(solver='lbfgs', max_iter=1000)

# Aplicando o KFold
cv = KFold(n_splits=10, shuffle=True)

# Aplicando a validação de cruzada
result = cross_validate(model, x, y, cv = cv)

# Visualizando a acurácia média
result['test_score'].mean()
````

````python
# Importando função de regressão logística
from sklearn.linear_model import LogisticRegression

# Importando função para separar dados
from sklearn.metrics import confusion_matrix

# Separando dados entre entrada e saída
x = diabetes.drop(columns=['class'])
y = diabetes['class']

# Separando dados entre dados de treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=100, test_size=0.25)

# Instânciando um objeto de regressão logística
model = LogisticRegression(solver='lbfgs', max_iter=1000)

# Treinando modelo
model.fit(x_train, y_train)

# Gerando os dados preditos
y_pred = model.predict(x_test)

# Matriz de confusão
cm = confusion_matrix(y_test, y_pred)

# Gerando um gráfico para a matriz de confusão
sns.heatmap(cm, fmt='d', annot=True, cmap='magma')
````
