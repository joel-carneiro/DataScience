# 1 - Regressão Linear

```r
# Criando variável independente
x <- seq(0, 100, 1)
x
```

```r
# Criando variável que iremos dependente
y <- (2 * x + 35) + rnorm(101, 0, 30)
y
```

```r
# Veremos agora um scatterplot das duas variáveis
plot(x = x, y = y)
```

```r
# Criando um modelo de regressão linear
linear_model <- lm(y ~ x)
```

```r
# Classe da regressão linear
class(linear_model)
```

```r
# Capturando os repectivos coeficientes
beta <- linear_model$coefficients[1]
alpha <- linear_model$coefficients[2]
```

$$ f(x) = ax + b$$

```r
# Predizendo valores
y_pred <- alpha * x + beta
```

```r
# Plotando a correlação de x e y 
plot(x = x, y = y, pch = 16)
lines(x, y_pred, lwd = 2, col = 'blue')
```

# 2 - Avançando com Exemplos

```r
age <- seq(0, 75, 1)
tmp <- 25 * (-age + rnorm(n = 76, mean = 100, sd = 10))

olds <- data.frame(age = age, tmp = tmp)

head(olds)
```

```r
plot(tmp ~ age, data = olds)
```

```r
linear_model <- lm(tmp ~ age, data = olds)
```

```r
linear_model
```

```r
beta <- linear_model$coefficients[1]
alpha <- linear_model$coefficients[2]

y_pred <- beta + alpha * olds$age
```

```r
plot(tmp ~ age, data = olds)
lines(y_pred, col = 'blue')
```
