# 1 - Classificação de Animais

```python
# Importando bibliotecas inicialmente necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ignorando possíveis avisos
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importando a base de dados
animals = pd.read_csv('/home/joeldspy/Modelos/datasets/animals.csv')
animals
```

```python
# Separando os dados entre entrada e saída
x = animals[['pelo_longo', 'perna_curta', 'late']]
y = animals['class']
```

```python
# Criando um modelo de classificação
from sklearn.svm import LinearSVC
model = LinearSVC()
```

```python
# Treinando o modelo
model.fit(x, y)
```

```python
# Predizendo um animal misterioso
rand_animal = [0, 1, 1]

model.predict([rand_animal])
```

```python
# Importando módulo para acurácia
from sklearn.metrics import accuracy_score
```

```python
# Adicionando a previsão do modeo
animals['predict'] = model.predict(x)
```

```python
# Calculando a porcentagem de acerto
sum(animals['class'] == animals['predict']) / len(animals) 
```

```python
# Criando alguns dados de treino
test = pd.DataFrame({
    'pelo_longo' : [1, 1, 0, 0, 1, 0],
    'perna_curta' : [1, 0, 1, 1, 0, 1],
    'late' : [1, 0, 0, 1, 1, 0],
    'class' : [1, 0, 0, 1, 0, 0]
})
```

```python
test.head()
```

```python
# Separando entre entrada e saída
x_test = test.iloc[:, :-1]
y_test = test.iloc[:, -1]
```

```python
# Predizendo valores
y_pred = model.predict(x_test)
y_pred
```

```python
# Calculando a Acurácia
accuracy_score(y_test, y_pred)
```

# 2 - Classificação de Vendas

```python
trackings = pd.read_csv('/home/joeldspy/Modelos/datasets/trackings.csv')
```

```python
trackings.head()
```

```python
trackings.shape
```

```python
x = trackings[['home', 'how_it_works', 'contact']]
y = trackings['bought']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=75)
```

```python
svc = LinearSVC()

svc.fit(x_train, y_train)

y_pred = svc.predict(x_test)

accuracy_score(y_test, y_pred)
```

# 3 - Projetos de Sites

```python
sites = pd.read_csv('/home/joeldspy/Modelos/datasets/sites.csv')
```

```python
sites.head()
```

```python
sites['finished'] = sites['unfinished'].map({0 : 1, 1 : 0})
```

```python
sites.head()
```

```python
sites.drop(['unfinished'], axis=1, inplace=True)
```

```python
sites.head()
```

```python
sns.scatterplot(x = 'expected_hours', y = 'price', data = sites)
```

```python
sns.scatterplot(x = 'expected_hours', y = 'price', data = sites, hue = 'finished')
```

```python
sns.relplot(x = 'expected_hours', y = 'price', data = sites, col = 'finished', hue='finished')
```

```python
x = sites[['expected_hours', 'price']]
y = sites['finished']

x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, random_state=75)

svc = LinearSVC()

svc.fit(x_train, y_train)

y_pred = svc.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)

print('Tamanho do treino: %d' % len(x_train))
print('Tamanho do teste: %d' % len(x_test))
print('Acurácia: %.2f%%' % (accuracy * 100))
```

```python
from sklearn.dummy import DummyClassifier

x = sites[['expected_hours', 'price']]
y = sites['finished']

x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, random_state=75)

dummy = DummyClassifier()

dummy.fit(x_train, y_train)

y_pred = dummy.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)

print('Tamanho do treino: %d' % len(x_train))
print('Tamanho do teste: %d' % len(x_test))
print('Acurácia: %.2f%%' % (accuracy * 100))
```
